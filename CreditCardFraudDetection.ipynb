```python
#importing libraries
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras import models,layers
from sklearn.model_selection import train_test_split
from tensorflow.keras.regularizers import l2
```

```python
dataset = pd.read_csv('creditcard.csv')
```

```python
dataset
```

```python
dataset.info() #analyzing data
```

```python
dataset.describe()
```

```python
dataset.isnull().sum() #checking for null values
```

```python
#standarizing time
Time_mean=dataset['Time'].mean(axis=0)
Time_std=dataset['Time'].std(axis=0)
dataset['Time']-=Time_mean
dataset['Time']/=Time_std
```

```python
#standarizing amount features
Amount_mean=dataset['Amount'].mean(axis=0)
Amount_std=dataset['Amount'].std(axis=0)
dataset['Amount']-=Amount_mean
dataset['Amount']/=Amount_std
```

```python
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
#Splitting data into train, test & val data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.2, random_state= 0)
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=0.3, random_state=0)
```

```python
#model creation
model=models.Sequential()
model.add(layers.Dense(6, kernel_regularizer=l2(0.01), activation='relu'))
model.add(layers.Dropout(0.5))
model.add(layers.Dense(6, kernel_regularizer=l2(0.01), activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))
model.compile(optimizer= 'adam', loss= 'binary_crossentropy', metrics= ['accuracy'])
model.fit(x_train, y_train, epochs=5, batch_size=32, validation_data=(x_val, y_val))
```

```python
#evaluation test dat
model.evaluate(x_test,y_test)
```
